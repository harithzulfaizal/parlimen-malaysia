```python
import re
import string 

with open('D:/User/Documents/.Parlimen Malaysia/AUMDRcompiled.txt', "r") as text_file:
        aumdr = text_file.read()

        
aumdr_split = re.split(r'PR', aumdr)

def remove_noise(data):
    aumdr_cleaned = []
    
    for i in range(len(data)):
        cleaned = re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+#]|[!*\(\),]|'\
                       '(?:%[0-9a-fA-F][0-9a-fA-F]))+','', data[i])
        aumdr_cleaned.append(cleaned.lower())
    
    return aumdr_cleaned

stopwords=['dan', 'atau', 'walaupun', 'bagaimana', 'ia', 'iaitu', 'adalah', 'itu', 'ini', 'bagi', 'yang', 'ada', 'begitu', 'dalam']

#from nltk.corpus import stopwords
#stop_words=stopwords.words('indonesian')

aumdr_cleaned=remove_noise(aumdr_split)

aumdr_cleaned2x=[]
for i in range(len(aumdr_cleaned)):
    wordz = ''.join([j for j in aumdr_cleaned[i] if j not in string.punctuation])
    aumdr_cleaned2x.append(wordz)
#print(aumdr_cleaned2x)

#print(aumdr_cleaned)

aumdr_cleaned2 = []
for i in aumdr_cleaned2x:
    #changing , to nested lists/arrays
    aumdr_cleaned2.append(i.replace(","," ").lower().split())
    
print(aumdr_cleaned2)
        

removed0=[]
for i in range(len(aumdr_cleaned2)):
    removed0.append(aumdr_cleaned2[i].pop(0))
    

tempaumdr=[]

for i in range(len(aumdr_cleaned2)):
    resultwords  = [word for word in aumdr_cleaned2[i] if word.lower() not in stopwords]
    result = ' '.join(resultwords)
    tempaumdr.append(result)
    
aumdr_cleaned3=[]
for i in tempaumdr:
    # Replacing "," , converting to lower and then splitting
    aumdr_cleaned3.append(i.replace(","," ").lower().split())

#print(aumdr_cleaned3)
print(sum(len(x) for x in aumdr_cleaned3))

output=[]
for elem in aumdr_cleaned3:
    output.extend(elem)
#print(output)

#removing punctuations
aumdr=[]
for i in range(len(output)):
    sentence = ''.join([letter for letter in output[i] if letter not in string.punctuation])
    aumdr.append(sentence)

#test_list = [i for i in aumdr if i] 
#print(test_list)

while("" in aumdr) : 
    aumdr.remove("") 
    
print(aumdr)


#print(removed0)
#print(aumdr_cleaned2[1][7])
```
